```python
train = pd.read_csv('/content/drive/MyDrive/SympactAI_Testing_Data/train_data.csv')
train = train.set_index(train.columns[0])
train.index.names = ['']
# train = train.set_index(train.columns[0])
# train.index.names = ['']

# list_ticker = train["tic"].unique().tolist()
# list_date = list(pd.date_range(train['date'].min(),train['date'].max()).astype(str))
# combination = list(itertools.product(list_date,list_ticker))

# train_full = pd.DataFrame(combination,columns=["date","tic"]).merge(train,on=["date","tic"],how="left")
# train_full = train_full[train_full['date'].isin(train['date'])]
# train_full = train_full.sort_values(['date','tic'])

# train = train.fillna(0)
```

```python
stock_dimension = len(train.tic.unique())
state_space = 1 + 2*stock_dimension + len(INDICATORS)*stock_dimension
print(f"Stock Dimension: {stock_dimension}, State Space: {state_space}")
train.head(20)
```

```python
buy_cost_list = sell_cost_list = [0.001] * stock_dimension
num_stock_shares = [0] * stock_dimension

env_kwargs = {
    "hmax": 100,
    "initial_amount": 1000000,
    "num_stock_shares": num_stock_shares,
    "buy_cost_pct": buy_cost_list,
    "sell_cost_pct": sell_cost_list,
    "state_space": state_space,
    "stock_dim": stock_dimension,
    "tech_indicator_list": INDICATORS,
    "action_space": stock_dimension,
    "reward_scaling": 1e-4
}


e_train_gym = StockTradingEnv(df = train, **env_kwargs)
```

```python
env_train, _ = e_train_gym.get_sb_env()
print(type(env_train))
```

```python
agent = DRLAgent(env = env_train)

# Set the corresponding values to 'True' for the algorithms that you want to use
if_using_a2c = True
if_using_ddpg = True
if_using_ppo = True
if_using_td3 = True
if_using_sac = True
```

```python
agent = DRLAgent(env = env_train)
model_a2c = agent.get_model("a2c")

if if_using_a2c:
  # set up logger
  tmp_path = '/content/drive/MyDrive/SympactAI_Testing_Data/a2c_ppo_ddpg_sac_td3_logs/a2c'
  new_logger_a2c = configure(tmp_path, ["stdout", "csv", "tensorboard"])
  # Set new logger
  model_a2c.set_logger(new_logger_a2c)
```

```python
trained_a2c = agent.train_model(model=model_a2c,
                             tb_log_name='a2c',
                             total_timesteps=10000) if if_using_a2c else None
```

```python
trained_a2c.save("/content/drive/MyDrive/SympactAI_Testing_Data/trained_models/agent_a2c_10000") if if_using_a2c else None
```
